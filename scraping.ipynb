```python
import requests

url = 'https://www.moneycontrol.com/marketmarkets/indian-indices/?classic=true'

r = requests.get(url)

print(r.status_code)
```

```python
from bs4 import BeautifulSoup
```

```python
soup  = BeautifulSoup(r.text,'html')
```

```python
body = soup.body
```

```python
main_div = body.find('div',class_ = 'main')
```

```python
title_of_page = main_div.find('div',class_='FL b_20').div.string
```

```python
title_of_page
```

```python
tables = main_div.find_all('table',class_ = 'tbldata14 bdrtpg')
```

```python
headers_columns = tables[0].find_all('th')
```

```python
header_column_names = []
for header in headers_columns:
    header_column_names.append(header.string)
header_column_names
```

```python
# getting data
def get_column_data(row):
    return list(map(lambda x:x.string,row))

for row in tables[1:]:
    print(get_column_data(row.find_all('td')))
```
